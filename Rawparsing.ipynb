```python
from rawparser import *
import numpy as np
```

```python
rp = RawParser ('test-meas.dat')
```

```python
rp.ParseData()
```

```python
rp.SaveData('test-meas.h5')
```

```python
sdata = rp.sync
sdata.shape
```

```python
data = np.squeeze (rp.data)
data = np.transpose (data, (0, 2, 3, 4, 1, 5))
```

```python
img = np.zeros (data.shape,dtype=np.complex)
for echo in range (0,data.shape[3]):
    for rxch in range (0, data.shape[4]):
        for txch in range (0, data.shape[5]):
            img[:, :, :, echo, rxch, txch] = np.fft.fftshift(np.fft.ifftn(np.fft.fftshift(data[:, :, :, echo, rxch, txch])));
            
```

```python
rp.SaveData('meas.h5')
```
